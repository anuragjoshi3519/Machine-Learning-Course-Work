```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
import math
```

```python
iris = datasets.load_iris()
X = iris.data  
Y = iris.target
```

```python
cov_mat = np.cov(X.T)
```

```python
inv_cov = np.linalg.inv(cov_mat)
```

```python
x = []
for i in range(X.shape[1]):
    x.append(np.mean(X[:,i]))
```

```python
print(x)
```

```python
X_diff = X - x
```

```python
X_diffT = X_diff.T
```

```python
print(X_diff.shape)
print(X_diffT.shape)
print(inv_cov.shape)
```

```python
y = []
for i in range(X.shape[0]):
    temp = np.dot(X_diff[i],inv_cov)
    #temp =temp.reshape((4,1))
    temp = np.dot(X_diffT[:,i],temp)
    y.append(float(temp))
```

```python
print(y)
```

```python
outliers = []
std = np.std(y)
k = 3. * std 
m = np.mean(y)
up_t = m + k
low_t = m - k
outliers = []
for i in range(len(y)):
    if (y[i] >= up_t) or (y[i] <= low_t):
        outliers.append(i)  # index of the outlier
```

```python
print(outliers)
```
