## 3D Surface Plots

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
```

```python
#meshgrid

a=[1,2,3]
b=[4,5,6]

a,b=np.meshgrid(a,b)
print(a)
print(b)
```

```python
A=np.arange(-4,4,0.2)
B=A
A,B=np.meshgrid(A,B)
print(A)
print(B)
```

```python
m=A.shape[0]
n=A.shape[1]

for i in range(m):
    for j in range(n):
        print("({},{})".format(A[i,j],B[i,j]),end=' ')
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(A,B,A**2+B**2,cmap='Blues')
plt.show()
```

```python
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.contour(A,B,A**2+B**2)
plt.show()
```
