## K Means clustering generally fails as a clustering model when data needs complex non-linear classiers such as below 

```python
from sklearn.datasets import make_circles,make_moons
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1,y1=make_moons(n_samples=500,noise=0.1)
X2,y2=make_circles(n_samples=500,noise=0.02)
```

```python
plt.subplot(121)
plt.scatter(X1[:,0],X1[:,1],c=y1)
plt.title('Moons')
plt.subplot(122)
plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.title('Circles')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km1=KMeans(n_clusters=2)
km1.fit(X1)
pred1=km1.labels_
centers1=km1.cluster_centers_
plt.scatter(X1[:,0],X1[:,1],c=pred1)
plt.scatter(centers1[:,0],centers1[:,1],c='red',marker='*')
plt.title('Moons Prediction')
plt.show()
```

```python
km2=KMeans(n_clusters=2)
km2.fit(X1)
pred2=km2.labels_
centers2=km2.cluster_centers_
plt.scatter(X2[:,0],X2[:,1],c=pred2)
plt.scatter(centers2[:,0],centers2[:,1],c='red',marker='*')
plt.title('Circles Prediction')
plt.show()
```
