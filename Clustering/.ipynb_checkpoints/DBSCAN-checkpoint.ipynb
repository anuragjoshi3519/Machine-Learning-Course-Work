## Density Based Spatial Clustering of Applications with Noise

```python
from sklearn.datasets import make_circles,make_moons
import matplotlib.pyplot as plt
```

```python
X1,y1=make_moons(n_samples=500,noise=0.1)
X2,y2=make_circles(n_samples=500,noise=0.01)
```

```python
plt.figure(figsize=(16,6))
plt.subplot(121)
plt.scatter(X1[:,0],X1[:,1],c=y1)
plt.title('Moons')
plt.subplot(122)
plt.scatter(X2[:,0],X2[:,1],c=y2)
plt.title('Circles')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db1=DBSCAN(eps=0.25,min_samples=17)
db1.fit(X1)
pred1=db1.labels_
plt.scatter(X1[:,0],X1[:,1],c=pred1)
plt.show()
```

```python
db2=DBSCAN(eps=0.08,min_samples=2)
db2.fit(X2)
pred2=db2.labels_
plt.scatter(X2[:,0],X2[:,1],c=pred2)
plt.show()
print(pred2)
```
