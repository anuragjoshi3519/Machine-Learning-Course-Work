```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X,y=make_blobs(n_samples=800,n_features=2,centers=5,random_state=3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
kmc=KMeans(n_clusters=5)
```

```python
kmc.fit(X)
```

```python
pred=kmc.labels_
print(pred)     # each value shows each data point belongs to which cluster
```

```python
centers=kmc.cluster_centers_
print(centers)         ## Centers of the clusters
```

```python
## Plotting the clusters with their centers

plt.scatter(X[:,0],X[:,1],c=pred)
plt.scatter(centers[:,0],centers[:,1],color='black',marker='*')
plt.show()
```
