```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

#### Creating Own Classification Dataset

```python
mean1=[2,4]
mean2=[6,8]
cov1=[[1,0.3],[0.3,1]]
cov2=[[1.3,0.2],[0.2,1.2]]
dist1=np.random.multivariate_normal(mean1,cov1,400)
dist2=np.random.multivariate_normal(mean2,cov2,400)
X=np.vstack((dist1,dist2))
Y=np.zeros(X.shape[0])
Y[X.shape[0]//2:]=1

plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
def distance(xQuery,x):
    return np.sqrt(sum((x-xQuery)**2))

def knn(xQuery,X,Y,k):
    distVector=[]
    for i in range(X.shape[0]):
        d=distance(xQuery,X[i])
        distVector.append((d,Y[i]))

    distVector=sorted(distVector)
    distVector=np.array(distVector)
    distVector=distVector[:k]
    distVector=distVector[:,1]
    
    U=np.unique(distVector,return_counts=True)
    index=U[1].argmax()
    return U,U[0][index]
```

```python
x=[4,5]
plt.scatter(X[:,0],X[:,1],c=Y)
plt.scatter(x[0],x[1],c='red')
plt.show()
```

```python
p,predClass=knn(x,X,Y,7)
print(p)
print(predClass)
```
