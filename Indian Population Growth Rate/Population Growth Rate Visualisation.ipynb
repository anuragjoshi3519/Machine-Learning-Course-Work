```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Population.csv")
dataX=df.values
dataY=df.columns
```

# Data Preprocessing

```python
y=[int(i) for i in dataY]

x1=dataX[0,:]  # World Population data
x2=dataX[1,:]  # Indian Population data
x1=[''.join(i.split(' ')) for i in x1]
x2=[''.join(i.split(' ')) for i in x2]

x1=list(map(int,x1))
x2=list(map(int,x2))

#x1=(x1 - np.mean(x1))/np.std(x1)
#x2=(x2 - np.mean(x2))/np.std(x2)
```

# Visualisation

```python
plt.figure(figsize=[15,15])
plt.subplot(221)
plt.scatter(x1,y,label='World Population',color='r')
plt.legend()
plt.subplot(222)
plt.scatter(x2,y,label='Indian Population',color='g')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
lr1=LinearRegression()
lr2=LinearRegression()
```

```python
y_world=np.array(y).reshape(-1,1)
x_world=np.array(x1).reshape(-1,1)
lr1.fit(y_world,x_world)

i1=lr1.coef_
b1=lr1.intercept_
```

```python
y_india=np.array(y).reshape(-1,1)
x_india=np.array(x2).reshape(-1,1)
lr2.fit(y_india,x_india)

i2=lr2.coef_
b2=lr2.intercept_
```

```python
year_world=np.arange(1950,2030)
population_world=year_world*i1 +b1
plt.figure(figsize=[15,15])
plt.subplot(221)
plt.plot(population_world.reshape(-1,),year_world,label='World Population',color='r')
plt.scatter(x1,y,label='World Population',color='r')
plt.legend()

year_india=np.arange(1950,2030)
population_india=year_india*i2 +b2
plt.figure(figsize=[15,15])
plt.subplot(221)
plt.plot(population_india.reshape(-1,),year_india,label='Indian Population',color='r')
plt.scatter(x2,y,label='Indian Population',color='g')
plt.legend()

plt.show()
```
