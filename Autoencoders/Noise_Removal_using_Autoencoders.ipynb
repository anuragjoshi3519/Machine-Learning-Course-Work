```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Conv2DTranspose, LeakyReLU, Reshape, Input, GaussianNoise
from tensorflow.keras.optimizers import Adam
```

```python
(XTrain,_), (XTest,_)= mnist.load_data()
print(XTrain.shape,XTest.shape)
```

```python
## Reshaping images
XTrain = XTrain.reshape((*XTrain.shape,1))
XTest = XTest.reshape((*XTest.shape,1))

## Normalizing Images
XTrain=XTrain/255.
XTest=XTest/255.

print(XTrain.shape)
```

### Adding Gaussian Noise in Training data

```python
XTrain_Noisey = XTrain + 0.22*np.random.randn(*XTrain.shape)
```

```python
plt.imshow(XTrain_Noisey[0].reshape(28,28))
```

# Autoencoder Architecture

### Encoder

```python
tf.random.set_seed(101)

encoder = Sequential()

encoder.add(Conv2D(32,kernel_size=(3,3),padding='same',strides=(2,2),input_shape=(28,28,1)))
encoder.add(LeakyReLU())
encoder.add(Conv2D(64,kernel_size=(3,3),padding='same',strides=(2,2)))
encoder.add(LeakyReLU())
encoder.add(Flatten())
encoder.add(Dense(128,activation='relu'))
encoder.add(Dense(64,activation='relu'))

encoder.summary()
```

```python
decoder = Sequential()

decoder.add(Dense(128,activation='relu',input_shape=(64,)))
decoder.add(Dense(3136,activation='relu'))
decoder.add(Reshape((7,7,64)))
decoder.add(Conv2DTranspose(32,kernel_size=(3,3),padding='same',strides=(2,2)))
decoder.add(LeakyReLU())
decoder.add(Conv2DTranspose(1,kernel_size=(3,3),padding='same',strides=(2,2),activation='sigmoid'))
decoder.summary()
```

```python
inputs = Input(shape=(28,28,1))
x = encoder(inputs)
outputs = decoder(x)

autoencoder = Model(inputs=inputs,outputs=[outputs])
autoencoder.summary()
```

```python
autoencoder.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
autoencoder.fit(XTrain_Noisey,XTrain,batch_size=256,epochs=5)
```


## Autoencoder predictions

```python
fig = plt.figure(figsize=(12,12))
idx = 1
n = 8
print('Original Image\t\t     Noisey Image\t\tReconstructed Image ')
for _ in range(n):
    ax=fig.add_subplot(n,3,idx)
    ax.imshow(XTest[idx].reshape(28,28))
    
    noisey_image = XTest[idx]+0.22*np.random.randn(*XTest[idx].shape)
    out = autoencoder.predict(np.expand_dims(noisey_image,axis=0))
    
    ax1=fig.add_subplot(n,3,idx+1)
    ax1.imshow((noisey_image).reshape(28,28))
    
    ax2=fig.add_subplot(n,3,idx+2)
    ax2.imshow(out[0].reshape(28,28))
    
    idx+=3
```
