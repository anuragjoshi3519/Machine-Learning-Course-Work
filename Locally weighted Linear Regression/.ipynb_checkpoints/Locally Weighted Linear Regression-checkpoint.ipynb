```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def readFile(name):
    return pd.read_csv(name)

X=np.array(readFile("./weightedX.csv"))
Y=np.array(readFile("./weightedY.csv"))
X=(X-np.mean(X))/np.std(X)
```

```python
plt.style.use('seaborn')
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
def getWMatrix(xQuery,X,tau):
    X=np.mat(X)
    W=np.mat(np.eye(X.shape[0]))
    for i in range(X.shape[0]):
        W[i,i]=np.exp(np.dot(X[i]-xQuery,(X[i]-xQuery).T)/(-2*tau*tau) )
    return W

W=getWMatrix(1,X,0.5)
print(W)
```

```python
def predict(xQuery,X,Y,tau):
    X=np.mat(X)
    Y=np.mat(Y)
    ones=np.ones((X.shape[0],1))
    X=np.hstack((X,ones))
    qx=np.mat([xQuery,1])
    W=getWMatrix(qx,X,tau)
    
    theta=np.linalg.pinv(X.T*W*X)*(X.T*W*Y)   # * operator works as matrix multiplication when used with matrices
    pred = qx*theta  # matrix multiplication
    return theta,pred
```

```python
theta,pred=predict(2,X,Y,0.5)
print(theta)
print(pred)
```

```python
print(X.shape,Y.shape,type(X))
```

# Visualisation

```python
def plotPrediction(X,Y,tau):
    '''
    tau is bandwidth(points within this bandwidth will
    be assigned weights greater than 0 and will mainly be used to
    predict the x_query point)
    If it is larger than the range of X then algo will
    work as normal linear regression. So it must be small wrt the range.
    Also it should not be too small, as there might be no neighbour points
    within the tau bandwidth
    '''
    x_test=np.linspace(-2,2)
    y_test=[]
    
    for qx in x_test:
        theta,pred=predict(qx,X,Y,tau)
        y_test.append(pred[0])
    plt.scatter(x_test,y_test,color='red',label='Predicted Data')
    plt.title('For tau/bandwidth = {}'.format(tau))
    plt.scatter(X,Y,label='Training Data')
    plt.legend()
    plt.show()
```

```python
taus=[0.0001,0.2,0.5,1,5]
for tau in taus:
    plotPrediction(X,Y,tau)
```
