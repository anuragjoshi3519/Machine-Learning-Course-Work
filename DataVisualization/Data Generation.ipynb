# Regression data generation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
std=4
mean=40
x=(np.random.randn(1000)*std)+mean # data generated is in standard normal distribution
noise=np.random.randn(1000)

y= x*3 + 5 +noise # y=mx+c +noise
```

```python
plt.scatter(x,y)
plt.show()
```

# Classification data generation

```python
#class 1 data
mean1=np.array([2,4])
cov1=np.array([[1,0.5],[0.5,0.7]])
X1=np.random.multivariate_normal(mean1,cov1,300)

#class 2 data
mean2=np.array([-1,-3])
cov2=np.array([[1,0.8],[0.8,2]])
X2=np.random.multivariate_normal(mean2,cov2,400)
```

```python
print(X1.shape)  # class 1 is apple and class 2 is oranges, both have 2 features x1(size) and x2(sweetness)
print(X2.shape)
```

```python
#plot of data in Class 1
plt.scatter(X1[:,0],X1[:,1])
plt.title("Apples")
plt.xlabel("x1 (Size)")
plt.ylabel("x2 (Sweetness)")
plt.show()
```

```python
#plot of data in Class 2
plt.scatter(X2[:,0],X2[:,1])
plt.title("Oranges")
plt.xlabel("x1 (Size)")
plt.ylabel("x2 (Sweetness)")
plt.show()
```

```python
#combined
plt.scatter(X1[:,0],X1[:,1],label="Apples")
plt.scatter(X2[:,0],X2[:,1],label="Oranges")
plt.legend()
plt.title("Apples Vs Oranges")
plt.xlabel("x1 (Size)")
plt.ylabel("x2 (Sweetness)")
plt.show()
```

```python
#combining data
y=np.zeros(700)
y[:300]=1
```

```python
X=np.vstack([X1,X2])      #final generated data (having 2 features x1 and x2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Apples Vs Oranges")
plt.xlabel("x1 (Size)")
plt.ylabel("x2 (Sweetness)")
plt.show()
```

```python
#alternate way, but prefer the above one
for i in range(X.shape[0]):
    if y[i]==1:
        plt.scatter(X[i,0],X[i,1],color='r')
    else:
        plt.scatter(X[i,0],X[i,1],color='orange')
plt.show()     
```

# Data generation using sklearn.datasets

```python
from sklearn.datasets import *
```

```python
X,Y=make_regression?
```

```python
X,Y=make_regression(n_samples=100,n_features=1,n_informative=1,noise=12)
print(X.shape)
print(Y.shape)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,color='g')
plt.show()
```

```python
XC,YC= make_classification?
```

```python
XC,YC= make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,n_classes=3,n_clusters_per_class=1,class_sep=2,random_state=1)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(XC[:,0],XC[:,1],marker='^',c=YC)
plt.show()
```

```python
#make_blobs are more useful in generating data for clustering problems
XB,YB=make_blobs(n_samples=1000,n_features=2,centers=6,random_state=10) # random_state fix the generated data
```

```python
plt.scatter(XB[:,0],XB[:,1],c=YB)
plt.show()
```

```python
#make_moons are used to generate dataset which needs a non-linear classifier to separate different classes
XM,YM=make_moons?
```

```python
XM,YM=make_moons(n_samples=100, noise=0, random_state=5)
plt.scatter(XM[:,0],XM[:,1],c=YM)
```

```python
XM,YM=make_moons(n_samples=100, noise=0.2, random_state=5)
plt.scatter(XM[:,0],XM[:,1],c=YM)
```
