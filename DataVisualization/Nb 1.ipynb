```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Line Graph

```python
x=np.arange(10)
y=np.arange(10)**2
```

```python
plt.figure(figsize=(7,7))
plt.scatter(x,y,marker='^',color='g',label='Quadratic Function')
plt.legend()
plt.xlabel("values")
plt.ylabel("squares")
plt.title("Plot 1")
plt.show()
```

```python
x1=np.arange(10)
y1=np.arange(10)**3
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y,color='r',label="Quadratic Function")
plt.plot(x1,y1,color='g',label='Cubic Function')
plt.legend()
plt.show()
```

# Bar Graph

```python
heights=np.random.randint(1,10,8)
indices=np.arange(8)
```

```python
plt.figure(figsize=(10,6))
plt.bar(indices,heights,width=0.5,color='g',label="Gold price")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.bar(indices-0.25,heights,width=0.5,color='g',label="Gold")
plt.bar(indices,heights/2,width=0.5,color='b',label="Silver")
plt.legend()
plt.title("Prices")
plt.show()
```

# Pie Chart

```python
label=['A','B','C','D']
values=[20,40,60,50]
plt.pie(values,labels=label,radius=1,explode=(0,0,1,0),shadow=True)
plt.show()
```
