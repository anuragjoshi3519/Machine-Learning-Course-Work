```python
import math
n= int(input())
print("Prime numbers upto {} are:".format(n))
for i in range(2,n+1):
    flag=0
    for j in range(2,math.floor((math.sqrt(i)+1))):
        if i!=2 and i%j==0:
            flag=1
            break
    if flag==0:
        print(i)
```
