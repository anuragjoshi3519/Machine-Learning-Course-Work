```python
#!pip install gTTS
#!pip install pyttsx3
import pyttsx3
from gtts import gTTS
import os
from tkinter import *
import nltk
from nltk.stem.lancaster import LancasterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
stemmer = LancasterStemmer()

import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.np_utils import to_categorical
import random
```

```python
import json
with open('intents.json') as file:
    data = json.load(file)


labels = []
docs_x = []
docs_y = []

for intent in data['intents']:
    for pattern in intent['patterns']:
        docs_x.append(pattern)
        docs_y.append(intent["tag"])
        
    if intent['tag'] not in labels:
        labels.append(intent['tag'])
        
tfidf_vectorizer=TfidfVectorizer(ngram_range=(1,2))
tfidf_vectorizer.fit(docs_x)
X_train=tfidf_vectorizer.transform(docs_x).toarray()
#print(tfidf_vectorizer.get_feature_names())
#print(tfidf_vectorizer.inverse_transform(tv[25]))

labels = sorted(labels)

Y_train = []
out_empty = [0 for _ in range(len(labels))]

for i in range(len(docs_y)):
    output_row = out_empty[:]
    output_row[labels.index(docs_y[i])] = 1
    Y_train.append(output_row)
Y_train=np.array(Y_train)    
```

```python
model = Sequential()
model.add(Dense(8,input_shape=(len(X_train[0]),),activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(len(Y_train[0]),activation='softmax'))
model.compile(optimizer='adam',loss='categorical_crossentropy')
model.summary()
```

```python
model.fit(X_train,Y_train, epochs=800, batch_size=8)
```

```python
def text_to_speech1(mytext):
    myobj = gTTS(text=mytext, lang='en', slow=False) 
    myobj.save("message.mp3") 
    os.system("mpg321 message.mp3")
def text_to_speech2(mytext):  # espeak required
    engine = pyttsx3.init()
    engine.setProperty('rate', 150)
    engine.say(mytext) 
    engine.runAndWait()

def getReply(question):
    results = model.predict(tfidf_vectorizer.transform([question]).toarray()[0].reshape(1,-1))  ##bag_of_words(inp, words).reshape(1,-1)
    results_index = numpy.argmax(results)
    tag = labels[results_index]

    for tg in data["intents"]:
        if tg['tag'] == tag:
            responses = tg['responses']

    return random.choice(responses)

window = Tk()
window.title(" Bank Assistant ")
messages = Text(window)
messages.pack()

input_user = StringVar()
bot=StringVar()

Label(window, text=" Enter Message : ").pack(side=LEFT) 
input_field = Entry(window, text=input_user)
input_field.pack(side=BOTTOM, fill=X)

def Enter_pressed(event):
    input_get = input_field.get()
    bot.set(getReply(input_get))
    messages.insert(INSERT, 'You: %s\n\n' % input_get)
    messages.insert(INSERT, 'Bot: %s\n\n' % bot.get())
    input_user.set('')
    text_to_speech2(bot.get())
    return "break"

frame = Frame(window)
input_field.bind("<Return>", Enter_pressed)
frame.pack()

window.mainloop()
```
