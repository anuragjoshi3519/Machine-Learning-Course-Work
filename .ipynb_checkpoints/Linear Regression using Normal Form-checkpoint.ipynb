```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression 
```

```python
X,Y=make_regression(n_samples=500,n_features=1,n_informative=1,noise=1.6,random_state=3)
plt.scatter(X,Y)
plt.show()
```

```python
Y=Y.reshape((-1,1))
Y=np.mat(Y)
```

```python
ones=np.ones((X.shape[0],1))
X1=np.hstack((X,ones))
```

```python
def predict(X,theta):
    return np.dot(X,theta)

def normalForm(X,Y):           # inverse(transpose(X).X).(transpose(X).Y)
    first=np.dot(X.T,X)
    second=np.dot(X.T,Y)
    theta=np.dot(np.linalg.pinv(first),second)
    return theta
```

```python
theta=normalForm(X1,Y)
```

```python
predY=predict(X1,theta)
```

```python
plt.scatter(X,Y)
plt.plot(X1[:,0],predY,label="Predicted function",color='r')
plt.legend()
plt.show()
```
